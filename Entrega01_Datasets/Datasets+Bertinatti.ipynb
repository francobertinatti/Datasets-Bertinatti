```python
import pandas as pd
import os
```

# DATASETS : UBER

```python
# Obtener la ubicación del directorio actual del notebook
directorio_actual = os.path.abspath(os.getcwd())

# Definir la ruta relativa al archivo CSV desde el directorio actual
ruta_csv_relativa = 'Dataset_Uber/uber_data.csv'

# Combinar la ruta del directorio actual con la ruta relativa
ruta_csv = os.path.join(directorio_actual, ruta_csv_relativa)

# Leer el archivo CSV con Pandas
try:
    df1 = pd.read_csv(ruta_csv)
    print("Archivo CSV cargado exitosamente.")
# Captura error
except FileNotFoundError:
    print(f"No se pudo encontrar el archivo CSV en la ruta: {ruta_csv}")
except Exception as e:
    print(f"Ocurrió un error al cargar el archivo CSV: {str(e)}")

df1[['passenger_count','trip_distance','pickup_longitude','pickup_latitude','fare_amount','tip_amount']].head()
```
